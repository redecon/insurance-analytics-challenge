```python
# 8. Create a second version (to prove versioning works)
import pandas as pd
df = pd.read_csv('MachineLearningRating_v3.txt', sep='|', low_memory=False, on_bad_lines='skip')
df['DVC_Version'] = 'v2_with_marker'
df.to_csv('MachineLearningRating_v3.txt', sep='|', index=False)

!dvc add MachineLearningRating_v3.txt
!git add MachineLearningRating_v3.txt.dvc
!git commit -m "chore(dvc): create data version 2 (demo of versioning)"
!dvc push
print("Second version created and pushed!")
```
